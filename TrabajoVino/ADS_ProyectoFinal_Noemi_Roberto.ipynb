# Descripción del Dataset:

Descripción del Dataset:
Existes dos dataset uno de vino blanco y el otro de vino tinto, se van a unir mediante la columna tipo de vino, que sera creada para guradarse en un solo csv

El dataset comprende las sigueintes columnas:

1) fixed acidity (Acidez fija):  Ácidos no volátiles, principalmente ácido tartárico, málico y cítrico (aunque el cítrico también puede considerarse por separado).
    **Rol:** Proporciona estructura y frescura al vino. Ayuda a la conservación y estabiliza el color. Una acidez fija demasiado baja hace que el vino sea plano, mientras que demasiada puede hacerlo agresivo.
    - Valor ideal: 5-7 g/L (varía por tipo de vino)

2) volatile acidity (Acidez volátil): Medida de los ácidos que pueden evaporarse fácilmente, principalmente ácido acético (vinagre).
    **Rol:** En bajos niveles, puede contribuir a la complejidad aromática. Niveles altos son considerados un defecto, dando un olor y sabor a vinagre.
    - Límite máximo: 0.6-1.2 g/L (depende de legislación)

3) citric acid (Ácido cítrico): Uno de los ácidos fijos, presente en menor cantidad que el tartárico y málico.
    **Rol:** Aporta frescura y puede usarse para ajustar la acidez. También tiene un efecto antioxidante. En exceso puede dar un carácter artificial.
    
4) residual sugar (Azúcares residuales): Cantidad de azúcar que queda después de la fermentación.
    **Rol:** Determina la dulzura del vino. En vinos secos, debe ser muy baja (generalmente < 4 g/L). En vinos dulces, puede ser alta. También contribuye al cuerpo y textura.
    - Clasificación:
        Seco: < 4 g/L
        Semiseco: 4-12 g/L
        Dulce: > 45 g/L

5) chlorides (Cloruros): Salinidad, medida de la cantidad de sal en el vino.
    **Rol:** En niveles muy bajos, puede ser imperceptible. En niveles moderados, puede realzar la sensación de cuerpo. En exceso, da un sabor salado no deseado.
    - Límite: < 100 mg/L (vinos de calidad)

6) free sulfur dioxide (Dióxido de azufre libre):  Forma de SO2 no unida a otros compuestos.
   **Rol:** Actúa como antioxidante y antiséptico, previniendo la oxidación y el crecimiento de microorganismos no deseados. Demasiado puede causar olores desagradables (como fósforos quemados) y reacciones alérgicas.

7) total sulfur dioxide (Dióxido de azufre total) : Suma del SO2 libre y el unido a otros compuestos (como azúcares).
    **Rol:** Indicador de la cantidad total de conservante añadido. Regulado por ley para evitar niveles perjudiciales para la salud.

8) density (Densidad): Masa por unidad de volumen, relacionada con el contenido de alcohol, azúcar y otros sólidos.
    **Rol:** Indicador del cuerpo del vino. Vinos con más alcohol o azúcar residual son más densos. También se usa para controlar la fermentación.
    Valores típicos: 0.990-1.005 g/mL
  
9) pH: Medida de la acidez activa (no confundir con la acidez total). Escala logarítmica inversa: a menor pH, mayor acidez.
    **Rol:** Afecta la estabilidad microbiológica, la eficacia del SO2, el color y el sabor. Vinos con pH bajo (3-3.5) son más ácidos y frescos; pH alto (3.6-4) son más propensos a contaminaciones.
    Rango ideal: 3.0-3.8

10) sulphates (Sulfatos): Sales del ácido sulfúrico, principalmente sulfato de potasio.
    **Rol:** Pueden provenir de la adición de sulfitos. Contribuyen a la conservación y pueden afectar la sensación en boca (amargor). Niveles altos pueden ser perjudiciales.

11) alcohol (Alcohol): Etanol, producto principal de la fermentación alcohólica.
    **Rol:** Aporta cuerpo, dulzor relativo (aunque es seco), calor y textura. También es un conservante. Niveles altos pueden dar calor alcohólico y desequilibrar el vino.
    - Rangos:
        Blancos: 11-14% vol
        Tintos: 13-16% vol

12) quality (Calidad): Evaluación sensorial del vino, generalmente en una escala (por ejemplo, de 0 a 10).
    **Rol:** Es la variable objetivo que refleja la percepción global del vino. Depende de todas las variables anteriores y de otros factores (como el equilibrio entre ellas).

```python
# Manejo de outliers
# Método: IQR score (boxplot)

import pandas as pd

def outliers(df_cat_vinos, column):
    Q1 =  df_cat_vinos[column].quantile(0.25)
    Q3 =  df_cat_vinos[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return  df_cat_vinos[( df_cat_vinos[column] >= lower_bound) & (df_cat_vinos[column] <= upper_bound)]

outlier_cols = ['residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide']
for col in outlier_cols:
    df_vinos = outliers( df_cat_vinos, col)
```
